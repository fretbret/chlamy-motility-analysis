## Introduction
---


Mutation is the ultimate source of complex trait variation and plays a central role in driving evolution. While variation introduced by mutation increases genetic variation in a population, other evolutionary forces also affect the total genetic variation$.^{7}$ Genetic drift can shuffle genetic variation or reduce it due to random chance, while natural selection can decrease variation by favoring traits (and their underlying mutations) that are beneficial to individuals$.^{11}$ Observing how much variation is retained in a population can thus help to explore the effect of mutation on complex traits. Populations of natural isolate (NI) strains and mutation accumulation (MA) strains serve as complementary models for investigating this variation: NI strains have different genetic backgrounds and undergo mutation under natural selection, while MA strains start with an identical genetic background and undergo mutation under minimal selection$.^{7}$ By comparing these two populations, we can parse the role of selection in shaping trait variation, concurrently examining how mutation alone influences trait variation.

To explore the effect of mutation on complex traits, phototaxis in the model organism *Chlamydomonas reinhardtii* (*C. reinhardtii*) was studied. Phototaxis describes the ability to move in response to a light stimulus, which in *C. reinhardtii* is driven by two motile flagella and a sensory eyespot organelle. Phototaxis is important in optimizing light absorption for photosynthesis$.^{12}$ A protocol was developed to quantify this trait in NI and MA strains. Much of the research on phototaxis in *C. reinhardtii* has focused on the molecular and cellular mechanisms that regulate flagellar movement$,^{3,9,14}$ along with modelling the cells' movement$.^{1,6}$ While some genes contributing to flagellar function and sensory mechanisms are known (eg. SHF1, FAP2, TTLL9)$,^{5}$ how mutations in these genes affect the overall phototaxis phenotype, particularly in the context of genetic variation within different *C. reinhardtii* strains, is not yet fully understood. The lack of experimental validation of candidate genes, particularly within a broad range of strains, leaves a significant gap in our understanding of phototaxis. This is important to explore, as there is conservation of proteins across *Chlamydomonas* and human cilia, which has previously been exploited to isolate human genes related to ciliary dysfunctions$.^{8}$

From an evolutionary perspective, the question was thus: what is the correlation between the phototaxis phenotype and the number of mutations in MA strains that fall within candidate phototaxis genes? This allows determination of the average effect of hit mutations on phototaxis (ie. if mutations tend to increase or decrease phototaxis). An exploratory analysis of candidate genes can also be performed. The nature of the questions depends on which direction the data points, but ultimately, may highlight genes of relevance in a future variant mapping experiment, using recombination to map effects of mutations on phototaxis. Possibilities include tracking candidate genes with the most hit mutations, checking for trends in the same hit candidate genes across strains with extreme phenotypes, and examining whether certain organelles or processes having a stronger effect on phototaxis through annotation information.

For this analysis, several files are available.

1. The C. reinhardtii CC-4532 v6.1 genome assembly and annotation was downloaded from Phytozome. This folder contains several files of relevance. Note that some files have transcriptName instead of geneName, in which case the gene that each transcript belongs to can easily be determined by taking the root of the name.

- "CreinhardtiiCC_4532_707_v6.1.annotation_info" is a tab-delimited file summarizing annotation details from Phytozome. Relevant columns include the Phytozome gene locus name (used to denote genes), the transcript name (used to denote transcripts), the GO column (which lists Gene Ontology terms associated with each gene, representing predicted biological processes, cellular components, and molecular functions), and the Best-hit-clamy-defline (which contains the top BLASTP alignment result when each protein in this proteome is aligned back against a previous Chlamydomonas reference proteome). The latter two columns can be used to filter candidate genes.

- "CreinhardtiiCC_4532_707_v6.1.description.txt" is a tab-delimited file with functional descriptions of all predicted transcripts in the Chlamydomonas reinhardtii CC-4532 v6.1 genome. The first column contains the transcript IDs (e.g., Cre01.g000050_4532.1), and the second column provides brief descriptions based on predicted domains, homology to known proteins, or gene family annotations. These descriptions are useful for interpreting gene function and prioritizing candidates based on known or inferred biological roles. The latter column can be used to filter candidate genes.

- "CreinhardtiiCC_4532_707_v6.1.geneName.txt" is a tab-delimited file with gene name annotations for each transcript in the Chlamydomonas reinhardtii CC-4532 v6.1 genome. The first column lists the transcript IDs (e.g., Cre01.g000050_4532.1), and the second column provides the corresponding gene or protein name (e.g., RWP-RK domain transcription factor). These names are useful for mapping transcripts to common gene identifiers.

- "CreinhardtiiCC_4532_707_v6.1.gene.gff3.gz" contains gene annotation data in GFF3 format, including genomic coordinates for features such as genes, mRNAs, UTRs, and CDS. This file is useful for mapping mutations to specific genomic regions of the CC-4532 v6.1 reference genome.

2. The "Mutation_Location.txt" tab-delimited file, generated by Rob Ness and Eniolaye Balogun, contains mutation data from multiple MA strains, including those used in this study. It includes genomic coordinates, functional annotations (e.g., synonymous, nonsynonymous mutations), gene feature overlaps, and quality metrics for each mutation. This file can be used to check mutations against candidate genes to generate hits across strains.

3. The "fitness+mut+pheno.csv" file integrates three types of information for each MA line: relative fitness compared to its ancestor, the total number of mutations accumulated during the MA experiment, and phototaxis phenotype. The main phenotypic metric, scaled_mean_max, reflects the average of the maximum distance travelled by a strain across multiple protocol iterations, scaled by its ancestor's value (MA / ANC). The fitness and mutation count were generated through previous Ness lab experiments, and the data was provided by Eniolaye Balogun. The phenotypic information was generated from the protocol quantifying phototaxis.

4. The "Phototaxis_Candidiate_Genes.xlsx" file lists 247 genes identified through GO term searches on the Gene Ontology Resource using keywords such as "cilia," "movement," "motility," "locomotion," "eyespot," and "flagella." Genes with common names in the "Gene/product (bioentity_label)" column can be linked to formal Phytozome gene identifiers by merging with the geneName.txt file, which maps common names to transcript IDs. The file can therefore be used to filter candidate genes.

```python
%%time

# Load annotation file from Phytozome (tab-delimited)
path = "/research/projects/chlamydomonas/PhototacticRate_AS/Real_Analysis/Phytozome/PhytozomeV13/CreinhardtiiCC_4532_anno/v6.1/annotation/CreinhardtiiCC_4532_707_v6.1.annotation_info.txt"
anno_df = pd.read_csv(path, sep='\t')

# List of biological keywords related to phototaxis or motility
keywords = ['flagella', "flagellum", "cilia", "cilium", "eyespot", "motile",
            "phototransduction", "photoreceptor", "blue-light", "axoneme", "axonemal"]

# Dictionary mapping selected Gene Ontology (GO) terms to human-readable descriptions
GO_dict = {
    "GO:0007602": "Phototransduction",
    "GO:0009881": "Photoreceptor activity", 
    "GO:0009637": "Response to blue light",
    "GO:0009583": "Detection of light stimulus",
    "GO:0009416": "Response to light stimulus",
    "GO:0050953": "Sensory perception of light",
    "GO:0001539": "Cilium or flagellum-dependent cell motility",
    "GO:0031514": "Motile cilium",
    "GO:0035082": "Axoneme assembly",
    "GO:0005930": "Axoneme",
    "GO:0030992": "intraciliary transport particle B",
    "GO:0009581": "Detection of external stimulus"
}

# Initialize list to store rows that match keywords or GO terms
matched_rows = []

# Iterate through each row in the annotation file
for _, row in anno_df.iterrows():
    source = None

    # Check if any keyword is found in the Chlamydomonas defline column
    if isinstance(row['Best-hit-clamy-defline'], str):
        for kw in keywords:
            if kw.lower() in row['Best-hit-clamy-defline'].lower():
                source = kw  # Store the keyword as the match source

    # If no keyword match, check for GO term matches
    if source is None and isinstance(row['GO'], str):
        go_terms = row['GO'].split()
        for go in go_terms:
            if go in GO_dict:
                source = GO_dict[go]  # Store the readable GO term as the match source
                break

    # If a match was found, append the row to the result list with source tag
    if source:
        new_row = row.copy()
        new_row['annofile_match_source'] = source
        matched_rows.append(new_row)

# Convert the matched results into a DataFrame
filtered_anno = pd.DataFrame(matched_rows)

# Keep only relevant columns and rename the GO column
filtered_anno = filtered_anno[['locusName', 'transcriptName', 'Best-hit-clamy-defline', 'GO', 'annofile_match_source']]
filtered_anno = filtered_anno.rename(columns={'GO': 'annofile_GO'})

# Display column names and sample data
print(filtered_anno.columns)
print(filtered_anno[['locusName', 'transcriptName', 'Best-hit-clamy-defline', 'annofile_GO', 'annofile_match_source']].head())
print(f"\nTotal matched genes: {len(filtered_anno)}\n")

# Count how many times each keyword or GO description was matched
source_counts = Counter(filtered_anno['annofile_match_source'])

# Print summary counts
print("\nMatch counts by source:")
for source, count in source_counts.items():
    print(f"{source}: {count}")
```

The counts for the transcript matches that each keyword/GO term resulted in are also highlighted, with flagella having the most hits (852), for a total of 953 transcripts.

### Annotation - GFF File

This file will provide genomic coordinates. The GFF was filtered for "gene" annotation types, since the other components are located within "gene". There are 16,883 gene annotations.

```python
# Define a function to parse a GFF3 gene annotation file
def parse_gff3_gene_info(gff_path):
    records = []  # Initialize an empty list to store parsed gene entries

    # Open the GFF3 file in text mode using gzip
    with gzip.open(gff_path, 'rt') as f:
        for line in f:
            if line.startswith('#'):
                continue  # Skip header/comment lines

            cols = line.strip().split('\t')  # Split the line by tabs
            if len(cols) != 9:
                continue  # Skip lines that do not have exactly 9 columns

            # Unpack the 9 GFF fields into separate variables
            chrom, source, feature_type, start, end, score, strand, phase, attributes = cols

            # Filter for relevant feature types (genes, mRNAs, CDS, UTRs)
            if feature_type in {'gene', 'mRNA', 'CDS', 'five_prime_UTR', 'three_prime_UTR'}:
                attr_dict = {}  # Dictionary to hold attribute key-value pairs
                attribute_items = attributes.split(';')  # Split the attribute field by semicolons

                for item in attribute_items:
                    if '=' in item:
                        key_value = item.split('=', 1)  # Split only on the first '='
                        if len(key_value) == 2:
                            key = key_value[0]
                            value = key_value[1]
                            attr_dict[key] = value  # Store in attribute dictionary

                # Try to extract gene identifier (ID or Parent), fallback to 'unknown' if missing
                geneName = attr_dict.get('ID') or attr_dict.get('Parent') or 'unknown'

                # Append a dictionary representing the gene/feature to the records list
                records.append({
                    'chrom': chrom,
                    'start': int(start),
                    'end': int(end),
                    'type': feature_type,
                    'strand': strand,
                    'geneName': geneName
                })

    # Return the list of records as a pandas DataFrame
    return pd.DataFrame(records)

path = "/research/projects/chlamydomonas/PhototacticRate_AS/Real_Analysis/Phytozome/PhytozomeV13/CreinhardtiiCC_4532_anno/v6.1/annotation/CreinhardtiiCC_4532_707_v6.1.gene.gff3.gz"
gff_df = parse_gff3_gene_info(path)  # Load parsed GFF3 data into DataFrame

print(len(gff_df))  # Print total number of entries in the GFF DataFrame

# Filter DataFrame to keep only rows where the feature type is 'gene'
filtered_gff_df = gff_df[gff_df['type'] == 'gene']

# Configure pandas to show full row/column content without truncation
with pd.option_context(
    'display.max_columns', None,          # Show all columns
    'display.max_rows', 100,              # Show up to 100 rows
    'display.width', None,                # Allow wide display
    'display.max_colwidth', None,         # Do not truncate long column entries
    'display.expand_frame_repr', False    # Keep rows side-by-side in single frame
):
    print(filtered_gff_df.iloc[:50])  # Print the first 50 filtered gene entries

print(len(filtered_gff_df))  # Print number of gene entries
```

### Count Mutations In/Proximal to Candidate Genes

This cell looks at how many mutations from the MA mutation info file fall into specific candidate genes. It tracks whether some genes have multiple mutations, and if they do, then which strains these mutations come from. It also contains the keyword that was used to isolate that gene.

```python
# Normalize chromosome labels in both dataframes by extracting chromosome numbers as floats
# Example: 'chromosome_01' → 1.0
mutations_df['norm_chrom'] = mutations_df['chromosome'].str.extract(r'(\d+)').astype(float)
candidate_genes['norm_chrom'] = candidate_genes['chrom'].str.extract(r'(\d+)').astype(float)

# Expand gene boundaries to ±10 bp to allow for slight position mismatches
candidate_genes['prox_start'] = candidate_genes['start'] - 10
candidate_genes['prox_end'] = candidate_genes['end'] + 10

# Initialize a list to store records of gene hits from mutations
gene_hit_records = []

# Loop through each mutation
for _, mut in mutations_df.iterrows():
    chrom = mut['norm_chrom']   # Extract normalized chromosome for this mutation
    pos = mut['position']       # Extract mutation position
    strain = mut['mutant_sample']  # Extract the strain the mutation is from

    # Filter candidate genes on the same chromosome
    chrom_genes = candidate_genes[candidate_genes['norm_chrom'] == chrom]

    # Check whether the mutation falls within any gene's start-end region
    for _, gene in chrom_genes.iterrows():
        if gene['prox_start'] <= pos <= gene['prox_end']:
            gene_hit_records.append({
                'chrom': gene['chrom'],
                'locusName': gene['locusName'],
                'strain': strain
            })

# Convert collected gene hit records to a DataFrame
hits_df = pd.DataFrame(gene_hit_records)

# Group by chromosome and gene, count number of hits, and collect unique strains
gene_summary = (
    hits_df.groupby(['chrom', 'locusName'])
           .agg(
               n_hits=('strain', 'count'),                     # Count how many times each gene was hit
               strains=('strain', lambda x: sorted(set(x)))   # Collect unique strains as a list
           )
           .reset_index()
           .sort_values('n_hits', ascending=False)            # Sort genes by number of hits
)

# Merge gene_summary with gene annotations to include match sources from different files
gene_summary = gene_summary.merge(
    candidate_genes[['locusName', 'annofile_match_source', 'descfile_match_source', 'gofile_match_source']],
    on='locusName',
    how='left'
)

# Output total number of gene hits (summed across all genes)
print(f"\nTotal number of hit mutations: {gene_summary['n_hits'].sum()}")

# Using Markdown to format the string with bold "Table 2"
text = " **Table 2.** Total number of candidate gene hits (summed across all strains) from MA mutation info file. The columns correspond to the chromosome of the gene, the gene ID, the number of hit mutations, and the keyword used to isolate the candidate genes from each of the three gene sources (ie. the annotation info file, the description file, and the GO terms online resource)."
# Display the formatted text
display(Markdown(text))

gene_summary
```

This cell looks at how many mutations from the MA mutation info file fall into specific candidate genes. It tracks how many MA mutations were located in any candidate phototaxis gene, and sums them within each strain.

```python
# Normalize chromosome labels in both dataframes
def normalize_chrom_label(label):
    # If the label is a string starting with "chromosome_", extract the number and convert it to int
    # e.g., "chromosome_01" → 1
    if isinstance(label, str) and label.startswith("chromosome_"):
        return int(label.split("_")[1].lstrip("0"))
    return label  # Return as-is for non-chromosome labels (e.g., "mtMinus", "cpDNA")

# Expand gene boundaries to ±10 bp to allow for slight position mismatches
candidate_genes['prox_start'] = candidate_genes['start'] - 10
candidate_genes['prox_end'] = candidate_genes['end'] + 10

# Apply normalization to both mutations_df and candidate_genes
mutations_df['norm_chrom'] = mutations_df['chromosome'].apply(normalize_chrom_label)
candidate_genes['norm_chrom'] = candidate_genes['chrom'].apply(normalize_chrom_label)

# Dictionary to count how many mutations fall within candidate genes for each strain
hit_counter = defaultdict(int)

# Iterate through every mutation
for _, mut in mutations_df.iterrows():
    chrom = mut['norm_chrom']         # Normalized chromosome number
    pos = mut['position']             # Mutation position
    strain = mut['mutant_sample']     # Strain the mutation belongs to

    # Subset candidate genes on the same chromosome
    chrom_genes = candidate_genes[candidate_genes['norm_chrom'] == chrom]
    
    # Check if the mutation falls within any candidate gene's extended region
    for _, gene in chrom_genes.iterrows():
        if int(gene['prox_start']) <= int(pos) <= int(gene['prox_end']):
            hit_counter[strain] += 1   # Increment counter for this strain
            break  # Prevent double-counting if multiple overlapping genes are present

# Convert results into a DataFrame
counts_df = pd.DataFrame(list(hit_counter.items()), columns=['strain', 'n_gene_hits'])

# Output summary statistics
print(f"\nTotal strains with at least one mutation in a candidate gene: {len(counts_df)}")
print(f"\nTotal number of hit mutations: {counts_df['n_gene_hits'].sum()}")
# Using Markdown to format the string with bold "Table 3"
text = " **Table 3.** Total number of candidate gene hits (summed across all genes) from MA mutation info file. The columns correspond to the given MA strain and its total number of hit mutations."
# Display the formatted text
display(Markdown(text))
counts_df
```

This cell looks at how many synonymous mutations from the MA mutation info file fall into specific candidate genes. It tracks how many MA synonymous mutations were located in any candidate phototaxis gene, and sums them within each strain.

There are 4 synonymous mutations in phototaxis candidate genes.

```python
# This version focuses only on mutations with synonymous/non-synonymous classification

# Normalize chromosome labels in both dataframes
def normalize_chrom_label(label):
    # Extracts the numeric part of chromosome labels like "chromosome_01" -> 1
    if isinstance(label, str) and label.startswith("chromosome_"):
        return int(label.split("_")[1].lstrip("0"))
    return label

# Apply normalization to both mutations and candidate gene datasets
mutations_df['norm_chrom'] = mutations_df['chromosome'].apply(normalize_chrom_label)
candidate_genes['norm_chrom'] = candidate_genes['chrom'].apply(normalize_chrom_label)

# Expand gene boundaries to ±10 bp to allow for slight position mismatches
candidate_genes['prox_start'] = candidate_genes['start'] - 10
candidate_genes['prox_end'] = candidate_genes['end'] + 10

# Only retain mutations that are explicitly annotated as "synonymous"
filtered_mutations = mutations_df[mutations_df['Nonsyn_V_Syn'] == "synonymous"]

# Initialize a dictionary to count hits per strain
hit_counter = defaultdict(int)

# Iterate through each filtered mutation
for _, mut in filtered_mutations.iterrows():
    chrom = mut['norm_chrom']         # Normalized chromosome label
    pos = mut['position']             # Mutation position
    strain = mut['mutant_sample']     # Strain ID

    # Filter genes to only those on the same chromosome
    chrom_genes = candidate_genes[candidate_genes['norm_chrom'] == chrom]

    # Check if mutation falls within the gene's proximity window
    for _, gene in chrom_genes.iterrows():
        if int(gene['prox_start']) <= int(pos) <= int(gene['prox_end']):
            hit_counter[strain] += 1   # Count a hit for this strain
            break  # Prevent counting the same mutation multiple times for overlapping genes

# Convert the dictionary to a DataFrame
syn_df = pd.DataFrame(list(hit_counter.items()), columns=['strain', 'n_gene_hits'])

# Output summary statistics
print(f"\nTotal strains with at least one mutation in a candidate gene: {len(syn_df)}")
print(f"\nTotal number of hit mutations: {syn_df['n_gene_hits'].sum()}")
# Using Markdown to format the string with bold "Table 4"
text = " **Table 4.** Total number of candidate gene synonymous hits (summed across all genes) from MA mutation info file. The columns correspond to the given MA strain and its total number of hit synonymous mutations."
# Display the formatted text
display(Markdown(text))
syn_df
```

### Merge Number of Hit Mutations in Each Strain With the Phenotype Info for Each Strain

This cell adds the n_gene_hit count for the number of hit mutations in candidate genes per strain to the dataframe containing phototaxis phenotype information for each strain.

```python
# Define a function to standardize strain labels for matching
def convert_strain_label(sample):
    # Converts mutation sample labels like 'CC2344_1' into 'CC-2344 L1'
    # and 'CC2931_13' into 'CC-2931 L13' for compatibility with phenotype data
    if "CC2344" in sample:
        num = sample.split("_")[1]
        return f"CC-2344 L{num}"
    elif "CC2931" in sample:
        num = sample.split("_")[1]
        return f"CC-2931 L{num}"
    else:
        return sample  # Return unchanged if the pattern doesn't match

# Apply the conversion to the mutation strain labels in counts_df
counts_df['strain_match'] = counts_df['strain'].apply(convert_strain_label)

# Merge the phenotype and mutation count data using the cleaned strain labels
pheno_geno = pheno_df.merge(
    counts_df[['strain_match', 'n_gene_hits']],  # Only need matchable strain and hit count
    left_on='strain', right_on='strain_match',   # Match phenotypes to reformatted mutation strain names
    how='left'                                   # Keep all phenotype rows (even if no mutations found)
)

# Replace missing hit counts with 0 and ensure the column is integer-typed
pheno_geno['n_gene_hits'] = pheno_geno['n_gene_hits'].fillna(0).astype(int)

# Define strains to exclude (e.g., ancestors and mutator line which is an outlier)
excluded_strains = ["CC-2344 ANC", "CC-2931 ANC", "CC-2344 L1"]

# Create a filtered DataFrame excluding the specified strains
pheno_geno_no_anc = pheno_geno[~pheno_geno['strain'].isin(excluded_strains)].copy()

# Final DataFrame combining phenotype and mutation hit data
pheno_geno
```

## Discussion
---

The correlation between the scaled mean maximum distance and the total number of candidate gene mutations was weak and the OLS regression was not statistically significant (r = -0.103, p = 0.609). The results were similar for synonymous mutations (r = 0.086, p = 0.671). This suggests that while mutations in candidate genes might influence phototaxis, they do not appear to have a strong directional effect. This supports unpublished results generated in the Ness lab for the phototaxis phenotype data. The variation in phototaxis of each MA group was much greater than the variation seen in the respective ancestor, reaching a similar degree as the NI group. This means that strains tend to get faster as much as they get slower, and faster is not always beneficial (likely due to some tradeoff).

An interesting but isolated observation is that a strain with an extreme phenotype had a synonymous mutation in a candidate gene: CC-2344 L10 (rank = 3). Some strains with extreme phototaxis phenotypes, such as CC-2344 L4 (1 mutation), did not exhibit a large number of hit mutations. Meanwhile, CC-2931 L9, despite having a relatively large number of hit mutations (5), did not exhibit an extreme phenotype. These inconsistencies may suggest that other factors, such as gene interactions or additional unknown genes, might be influencing the phototaxis phenotype.

Another finding was that mutations in flagella-related genes were the most common across strains. The common names of the most mutated genes were FAP51, LF3, FAP343, FAP376, FBB19, FAP221, and FAP184 based on **Table 6** (FAP = flagellar-associated protein , LF = long-flagella, FBB = flagella basal body). This is consistent with previous work on *C. reinhardtii*, where flagella-related genes are often implicated in motility, but a holistic overview of their role in phototaxis is still under investigation. A study of LF3 showed that mutants had defective flagellar assembly and length regulation$.^{10}$ A different study using axonemal extracts identified FAP221 as part of a calcium-dependent complex that regulates flagella motility by interacting with calmodulin in high calcium concentrations$.^{4}$ FAP genes generally play key roles involved in the assembly and function of the axoneme in eukaryotic flagella, but the remaining FAP genes have not been experimentally validated$.^{2}$ Generally, FBB genes have been shown to lead to short cilia and a loss of dynein arms, indicating a role in flagellar function$.^{13}$

Explanations for the lack of significant findings could be the sample size, along with the fact that candidate genes were assembled using BLAST and GO annotations instead of strictly literature-based genes. Only 440 candidate genes had proper coordinate annotation information. From these, only 30 genes did not harbour any NaN values; these rows may represent candidate genes with more confidence, as they were annotated from all three data sources. Regardless, the sample size and selection methodology may not be representative of the full genetic diversity of *C. reinhardtii* and might explain the lack of hit synonymous mutations. Experimental variant mapping data can help ascertain which candidate genes truly contribute to photoaxis.
